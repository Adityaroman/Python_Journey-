```python
#Different types of Mode
'r'   open for reading (default)         
'w'   open for writing, truncating the file first(overwrite = means first delete the original data and add it new further)
'x'   create a new file and open it for writing()
'a'   open for writing, appending to the end of the file if it exists(append add new data along with existing data without deleting like write)
'b'   binary mode
't'   text mode (open by default)t is implecite
'+'   open a disk file for updating (reading and writing)if you wann do or use any two out of all above
```

```python
def check_for_line():
    word = "programing"
    data = True
    line_no = 1
    with open("Sample.text", "r") as f:
        while data:
            data = f.readline()
            if word in data:
                print(line_no)
                return line_no
            line_no += 1
    return -1

print(check_for_line())
```

```python
#OR
count = 0
with open("io.txt", "r")as f:
    data=f.read()
    print(data)

    num = data.split(",")
    for val in num:
        if(int(val) % 2==0):
            count +=1
print(count)
```
